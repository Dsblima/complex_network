```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = [5,10,25,50,100]
p = [(item*0.05) for item in list(range(1,21,1))]
```

```python
class ComplexNetwork:

    def erdo_reny(self,n,p):
        print(n,p) 
        for _ in range(20):
            er = nx.erdos_renyi_graph(n,p)                      
            degree_sequence = list(er.degree())
            nb_nodes = n
            nb_ar = len(er.edges())
            #Coeficiente
            cc = list(nx.clustering(er).values())
            self.calcularPropriedades(degree_sequence,nb_nodes,nb_ar,cc)
    
    def barabasi_albert(self,n,p):
        print(n,p) 
        for _ in range(20):
            er = nx.barabasi_albert_graph(n,p)                      
            degree_sequence = list(er.degree())
            nb_nodes = n
            nb_ar = len(er.edges())            
            self.calcularPropriedades(degree_sequence,nb_nodes,nb_ar,cc)
    
    def calcularPropriedades(self,degree_sequence,nb_nodes,nb_ar,cc):
        avg_degree = np.mean(np.array(degree_sequence)[:,1])
        med_degree = np.median(np.array(degree_sequence)[:,1])
        max_degree = max(np.array(degree_sequence)[:,1])
        min_degree = np.min(np.array(degree_sequence)[:,1])
        cc = np.mean(cc)
        print("Number of nodes : " + str(nb_nodes))
        print("Number of edges : " + str(nb_ar))
        print("Maximum degree : " + str(max_degree))
        print("Minimum degree : " + str(min_degree))
        print("Average degree : " + str(avg_degree))
        print("Median degree : " + str(med_degree))
        print("Clustering Coefficient : " + str(cc))
        print('\n')

    
    
```

```python
net = ComplexNetwork()
net.erdo_reny(25,0.5)
```
